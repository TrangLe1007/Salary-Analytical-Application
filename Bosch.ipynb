```python
import csv
import pandas as pd
import re
path = "C:/Users/Admin/Downloads/V7_VersionDescriptionReport_PMT_5_9_0_to_5_9_02.csv"
data = pd.read_csv(path) 
df=pd.DataFrame(data)
with open(path, 'r',encoding='utf-8-sig',errors='ignore') as infile:
    # this list will hold unique title
    description_lst = []
    # read input file into a dictionary
    results = csv.DictReader(infile)
    # writer = csv.writer(outfile)

    # write column headers to output file
    # writer.writerow(
        # ['Variant', 'Variant Version', 'Modules', 'Module Version', 'Description','ALM ID','Creation Date','Module ID'])
    for result in results:
        description = result.get('Description')
        description_lst.append(description) 
    # print(description_lst)
    id_lst=[]
    for link in description_lst:
         alm_id = re.findall(';id=(.+?)</p>', link)
        
         id_lst.append(alm_id)
    # print(id_lst) 

    #add data into each column
    # writer.writerow([
        #  result.get('Variant'),
        #  result.get('Variant Version'),
        #  result.get('Included Modules'),
        #  result.get('Module Version'),
        #  result.get('Description'),
        #  id_lst,
        #  result.get('Creation Date'),
        #  result.get('Module ID')
        # ])
df['ALM ID']=id_lst
print(df)
# df.to_csv('a.csv')
```
