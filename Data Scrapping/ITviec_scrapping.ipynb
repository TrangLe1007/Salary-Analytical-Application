```python
#Create empty list to store data
lst_salary=[]
lst_title=[]
lst_link=[]
lst_address=[]
lst_company_link=[]

url="https://itviec.com/it-jobs?page=1&source=search_job"

#create a loop to get data from mutiple pages
while True:
    #salary info of jobs are hidden behind a login thus need to identify what authorization technique is used
    session = "N5MRJF8QM5vIMavTZjGpVNwUEvNY3dC22JpgT%2F2xZOeiOJCcA2TwglX3iL66LRWFrTiaKNaMkK%2BfCdTLeAVP765d5SQEZn4PZf5e8y9Y%2Ba2iFMQoW%2Bj8un7aPr3FnqC13FKT%2Bm6yJ7xOd%2B48prHIfxq9kVs1P1C7ZM%2BUb8uq8%2FHguSqbvGQEEEeTdT6d63RxrWemo98Mvz5WNlC9%2FGOZQkFCOJCh9L6JEUe9SNYr6v6ThBAKvfMU1YzePVcD89FuLsrdtUHW8D3A532PYM74ZRZgq65dd0PqqBI5Yyk76FvU9wKGDIDOwJvhwfNIKmprxSySxV7bvM2CffzuMvW9dqDOJEV9Wh3v%2BOdq4ps3gt9drOVpdAkALe3sFVu5xqq64FNWtPjk6VJg4irdqC36YMmXGvvZoejb%2FuzB3O0PN%2FUPj2rlezHX%2BVPvDzay7zi1v2g6ghE5RmZlo0c1jjvALq1HYpQnTJt3Lo00c%2BEp0XI3TtrD2aYwHPi7GvEKF8CjjO88k10hE10LW8taHe6epE1KzKSYUNwi9WuQ61VPPbWQWMtkRg6H8CHs6w%3D%3D--m%2BHtTKnc9d%2Fcwldl--9isWji1wkVTPrwhzOL9DhA%3D%3D; path=/; HttpOnly; SameSite=Lax"   
   
    # attach ID into each request's cookie to trick the server that this request is made by a valid, logged-in user 
    response = requests.get(url,cookies={'_ITViec_session': session})
    data=response.text
    soup=bs(data,"html.parser")

    #scrape data
    salary=soup.findAll("div",{"class":"svg-icon__text"})
    for i in salary:
        lst_salary.append(i.text)
    box=soup.findAll("h2",{"class":"title"})
    for j in box:
        link=j.find("a").get("href")
        lst_link.append(link)
        title=j.find("a").text
        lst_title.append(title)
    address=soup.findAll("span",{"class":"text"})
    for f in address:
        lst_address.append(f.text)
    link_company=soup.findAll("div",{"class":"logo-wrapper"})
    for k in link_company:
        link_lst=k.find("a").get("href")
        lst_company_link.append(link_lst)

    url_tag=soup.find('a',{'rel':'next'})
    if url_tag["href"]!=None: 
        url="https://itviec.com/"+ url_tag["href"]
        print(url)
    else:
        break
```

```python
#create a dictionary to store list of data
database = {'Title_job': lst_title,
            'Job_link':lst_link,
            'Company_link':lst_company_link,
            'Address':lst_address,
            'Salary':lst_salary  
            }
#convert dictionary to dataframe by Pandas            
df = pd.DataFrame.from_dict(database, orient='index')

df = df.transpose()
#add column 'date' to save the day of scraping
df['date'] = pd.to_datetime('today').strftime("%m/%d/%Y")
# df['date']='8/20/2021'
#append dataframe into the exist CSV file

df.to_csv('IT_job_Database_2208.csv',encoding='utf-8-sig')
df.to_csv('IT_job_Database - Staging.csv', mode='a', index=True, header=False)

#check the first of 5 lines of table
df.head(5)
```
